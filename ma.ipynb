```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA

commodities = ['Maize', 'Wheat', 'Oats', 'Soybeans']
DATA_PATH = '/kaggle/input/processed-data' # Remove the /kaggle/input prefix for reproducing the results
Y_COLUMN = 'Sep'
```

```python
commodities_datasets = []
for commodity in commodities:
    dataset = pd.read_csv(f'{DATA_PATH}/full_{commodity}.csv')
    dataset['Date'] = pd.to_datetime(dataset['Date'])
    dataset = dataset[['Date', Y_COLUMN, "Prime"]]
    commodities_datasets.append(dataset)
```

```python
train_datasets = []
test_datasets = []
for i in range(len(commodities_datasets)):
    split_index = int(len(commodities_datasets[i]) * 0.8)
    train_df = commodities_datasets[i].iloc[:split_index]
    test_df = commodities_datasets[i].iloc[split_index:]
    train_datasets.append(train_df)
    test_datasets.append(test_df)
    assert len(commodities_datasets[i]) == len(train_df) + len(test_df)
```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Load your dataset (assuming it's in a pandas DataFrame 'df' with a 'value' column)
df = train_datasets[0]
data = df[Y_COLUMN]
```

```python
def fit_ma_model(data, max_q):
    best_aic = float('inf')
    best_order = None
    best_model = None
    
    for q in range(1, max_q+1):
        model = ARIMA(data, order=(0, 0, q))
        results = model.fit()

        print(f'Model MA({q}) - AIC: {results.aic}')

        # Choose the model with the lowest AIC
        if results.aic < best_aic:
            best_aic = results.aic
            best_order = q
            best_model = results

    print(f'\nBest model: MA({best_order}) - AIC: {best_aic}')
    return best_order, best_model
```

```python
q_values = []
for train_dataset in train_datasets:
    data = train_dataset[Y_COLUMN]
    best_order, best_model = fit_ma_model(data, max_q=15)  # You can adjust max_q as needed
    print(best_model.summary())

    residuals = best_model.resid
    plt.figure(figsize=(10, 4))
    plt.plot(residuals)
    plt.title('Residuals of the Best MA Model')
    plt.show()
    
    q_values.append(best_order)
```

# Forecasting with the best MA(q) models

```python
all_rmse_train = []  
all_rmse_test = [] 

for i, dataset in enumerate(commodities_datasets):
    best_ma_model = ARIMA(dataset[Y_COLUMN], order=(0, 0, q_values[i]))
    best_ma_model_fit = best_ma_model.fit()

    start = 1  # Start from the second observation
    end = len(dataset[Y_COLUMN]) - 1  # End index for predictions

    best_ma_model_fit_predictions = best_ma_model_fit.predict(start=start, end=end)

    # Slice the actual data to exclude the first observation
    actual_data_without_first = dataset[Y_COLUMN][1:]

    plt.figure(figsize=(20, 15))
    plt.plot(actual_data_without_first, label='Actual Data', color='blue')

    plt.plot(best_ma_model_fit_predictions, label=f'MA({q_values[i]}) Predictions', color='orange')

    plt.xlabel('Day')
    plt.ylabel('Contract Price')
    plt.title(f'Actual vs MA({q_values[i]}) Predicted Values (Excluding First Observation), commodity = {commodities[i]}')
    plt.legend()
    plt.xlim(0,)
    plt.show()
    print("--------------------")
    
    training_rmse = (((best_ma_model_fit_predictions[:-len(test_datasets[i])] - actual_data_without_first[:-len(test_datasets[i])]) ** 2).mean())**0.5
    test_rmse = (((best_ma_model_fit_predictions[-len(test_datasets[i]):] - actual_data_without_first[-len(test_datasets[i]):]) ** 2).mean())**0.5
    all_rmse_train.append(training_rmse)
    all_rmse_test.append(test_rmse)

```

```python
n_commodities = len(commodities)
ind = np.arange(n_commodities)  # the x locations for the commodities
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(10, 6))
train_bars = ax.bar(ind - width/2, all_rmse_train, width, label='Train RMSE loss')
test_bars = ax.bar(ind + width/2, all_rmse_test, width, label='Test RMSE loss')

ax.set_xlabel('Commodities')
ax.set_ylabel('RMSE')
ax.set_title('Comparison of RMSE loss across commodities, MA models')
ax.set_xticks(ind)
ax.set_xticklabels(commodities)
ax.legend()

# Function to add labels above the bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(train_bars)
add_labels(test_bars)
plt.tight_layout()
plt.show()
```
